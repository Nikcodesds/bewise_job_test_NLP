```python
# очищаем данные от знаков препинания, переводим в маленький регистр для улучшения поиска в дальнейшем.
def clean_up_data(data):
    """
    Cleans up the data from unnecessary signs
    """
   
    data = data.str.replace(r'+', '')
    data = data.str.replace(r'\d+', '')
    data = data.str.replace(r'%', '')
    data = data.str.replace(r',/./', '')
    data = data.str.lower()
    return data

df['text'] = clean_up_data(df['text'])
df

```

```python
# Просмотрим, какие разные слова в диалоге встречаются и как часто
df_manager = df.drop(df[df.role == 'client'].index)
df_client = df.drop(df[df.role == 'manager'].index)

manager_text = str('')
for i in range(len(df_manager)):
    manager_text += df_manager['text'].iloc[i] + ' '

blob = TextBlob(manager_text)

dict_words_manager = blob.word_counts
dict_words_manager

```

```python
# Весь диалог переведем в текст
whole_text = str('')
for i in range(len(df)):
    whole_text += df['text'].iloc[i] + '  '
print(whole_text)
```

```python
# Весь диалог от клиента переведем в текст
client_text = str('')
for i in range(len(df_client)):
    client_text += df_client['text'].iloc[i] + ' '
print(client_text)
```

```python
# Весь диалог от менеджера переведем в текст
manager_text = str('')
for i in range(len(df_manager)):
    manager_text += df_manager['text'].iloc[i] + ' '
print(manager_text)
```
